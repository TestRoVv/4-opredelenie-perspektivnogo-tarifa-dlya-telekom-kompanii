### –í—Ä–µ–º—è

–î–æ–±–∞–≤–∏–º —Å—Ç–æ–ª–±–µ—Ü —Å –º–µ—Å—è—Ü–∞–º–∏ –≤ —É–¥–æ–±–Ω–æ–º —Ñ–æ—Ä–º–∞—Ç–µ.

```python
def data_month(df,name_data_column):
    df[name_data_column] = pd.to_datetime(df[name_data_column], format='%Y %m %d') # –ø—Ä–µ–æ–±—Ä–∞–∑—É–µ–º —Ñ–æ—Ä–º–∞—Ç –≤ –Ω–æ—Ä–º–∞–ª—å–Ω—ã–π –≤–∏–¥
    time = []
    for i in df[name_data_column]:
        #day_week = i.strftime("%A")
        month = i.strftime("%B") # —Å–æ—Ö—Ä–∞–Ω—è–µ–º –º–µ—Å—è—Ü –æ—Ç–¥–µ–ª—å–Ω–æ
        year = i.strftime("%Y") # —Å–æ—Ö—Ä–∞–Ω—è–µ–º –≥–æ–¥ –æ—Ç–¥–µ–ª—å–Ω–æ      
        time.append([month,year]) 
    column1 = ['month', 'year']
    time_df = pd.DataFrame(data=time, columns=column1) #—Å–æ—Ö—Ä–∞–Ω—è–µ–º –≤—Ä–µ–º—è –æ—Ç–¥–µ–ª—å–Ω–æ
    df['month'] = time_df['month'] # —Å–æ—Ö—Ä–∞–Ω—è–µ–º –≤ df –º–µ—Å—è—Ü –æ—Ç–¥–µ–ª—å–Ω–æ, –≥–æ–¥ –Ω–µ –Ω—É–∂–µ–Ω –¥–ª—è —Ä–∞—Å—á–µ—Ç–æ–≤.
```

```python
data_month(calls_df,'call_date')
data_month(internet_df,'session_date')
data_month(messages_df,'message_date')
#data_month(users_df,'reg_date')
```

–°–¥–µ–ª–∞–µ–º —Ç–∞–±–ª–∏—Ü—É, —á—Ç–æ –±—ã –æ–Ω–∞ –æ—Ç–æ–±—Ä–∞–∂–∞–ª–∞ –∑–∞—Ç—Ä–∞—Ç—ã –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –ø–æ –º–µ—Å—è—Ü–∞–º –∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–≤–æ–Ω–∫–æ–≤, –∫–æ—Ç–æ—Ä–æ–µ –æ–Ω —Å–¥–µ–ª–∞–ª.

```python
calls = calls_df.pivot_table(index=['user_id','month'],values='duration',aggfunc=('count','sum'))
calls = calls.rename(columns={'count':'quantity_calls','sum':'sum_of_minutes'})
internet = internet_df.pivot_table(index=['user_id','month'],values='mb_used',aggfunc='sum')
internet = internet.rename(columns={'mb_used':'sum_mb'})
messages = messages_df.pivot_table(index=['user_id','month'],values='id',aggfunc='count')
messages = messages.rename(columns={'id':'quantity_messages'})
display(calls.head())
display(internet.head())
display(messages.head())
```

```python
def minutes_smart(num):
    denga = num-tariffs_df['minutes_included'][0]
    if denga<=0:
        return 0
    else:
        return denga*tariffs_df['rub_per_minute'][0]
    
def minutes_ultra(num):
    denga = num-tariffs_df['minutes_included'][1]
    if denga<=0:
        return 0
    else:
        return denga*tariffs_df['rub_per_minute'][1]
```

```python
def messages_smart(num):
    denga = num-tariffs_df['messages_included'][0]
    if denga<=0:
        return 0
    else:
        return denga*tariffs_df['rub_per_message'][0]
    
def messages_ultra(num):
    denga = num-tariffs_df['minutes_included'][1]
    if denga<=0:
        return 0
    else:
        return denga*tariffs_df['rub_per_minute'][1]
```

```python
def mb_per_smart(num):
    denga = num-tariffs_df['mb_per_month_included'][0]
    if denga<=0:
        return 0
    else:
        return (math.ceil(denga/1024))*tariffs_df['rub_per_gb'][0]
    
def mb_per_ultra(num):
    denga = num-tariffs_df['mb_per_month_included'][1]
    if denga<=0:
        return 0
    else:
        return (math.ceil(denga/1024))*tariffs_df['rub_per_gb'][1]
```

```python
ultra = median_combined_df.query('tariff == "ultra"')['price_median']
smart = median_combined_df.query('tariff == "smart"')['price_median']
```

–í—ã—è—Å–Ω–∏–º, –∫–∞–∫ —Ä–∞–∑–ª–∏—á–∞—é—Ç—Å—è –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –≤—ã–±–æ—Ä–æ–∫ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Ç–∞—Ä–∏—Ñ–∞ —Å–º–∞—Ä—Ç –∏ —Ç–∞—Ä–∏—Ñ–∞ —É–ª—å—Ç—Ä–∞.

```python
print('–î–∏—Å–ø–µ—Ä—Å–∏—è –≤—ã–±–æ—Ä–∫–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Ç–∞—Ä–∏—Ñ–∞ –°–º–∞—Ä—Ç:', smart.var(ddof=1))
print('–î–∏—Å–ø–µ—Ä—Å–∏—è –≤—ã–±–æ—Ä–∫–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Ç–∞—Ä–∏—Ñ–∞ –£–ª—å—Ç—Ä–∞:', ultra.var(ddof=1))
```

–ü—Ä–æ–≤–µ—Ä–∏–º –≥–∏–ø–æ—Ç–µ–∑—É, –∏—Å–ø–æ–ª—å–∑—É—è –≤—Å—Ç—Ä–æ–µ–Ω–Ω—É—é –≤ –±–∏–±–ª–∏–æ—Ç–µ–∫—É scipy —Ñ—É–Ω–∫—Ü–∏—é ttest_ind - –∫–æ—Ç–æ—Ä–∞—è –≤—ã—á–∏—Å–ª—è–µ—Ç —É—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ –∫—Ä–∏—Ç–µ—Ä–∏—è –°—Ç—å—é–¥–µ–Ω—Ç–∞ –¥–ª—è –¥–≤—É—Ö –Ω–µ–∑–∞–≤–∏—Å–∏–º—ã—Ö –≤—ã–±–æ—Ä–æ–∫. –¢.–∫. –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –¥–ª—è –≤—ã–±–æ—Ä–æ–∫ —Å–∏–ª—å–Ω–æ –æ—Ç–ª–∏—á–∞—é—Ç—Å—è, —Ç–æ –Ω–µ–±–æ—Ö–æ–¥–∏–º–æ —É—á–µ—Å—Ç—å —ç—Ç–æ –≤ —Ñ—É–Ω–∫—Ü–∏–∏, –¥–æ–±–∞–≤–∏–≤ –∞—Ä–≥—É–º–µ–Ω—Ç equal_var=False.

```python
alpha = .05

results = st.ttest_ind(
    ultra, 
    smart, 
    equal_var=False)

print('p-–∑–Ω–∞—á–µ–Ω–∏–µ:', results.pvalue)

if (results.pvalue < alpha):
    print("–û—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É")
else:
    print("–ù–µ –ø–æ–ª—É—á–∏–ª–æ—Å—å –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É")
```
